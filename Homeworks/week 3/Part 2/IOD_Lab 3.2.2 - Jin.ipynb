##### 4. Define Hypothesis

Let's assume if `RM` of house is more than 6, they are large house. Other houses are small. Price of houses increases when the `RM` increases.

**ANSWER:**

    H0: There is no difference between distribution of `Large House` & `Small House`

##### 5. Set alpha (Let alpha = 0.05)

First, we should specify our criterion for statistical significance. Traditionally, we allow for no more than a 1-in-20 chance of a spurious rejection of the null hypothesis:

```python
# ANSWER
alpha = 0.05
```

##### 6. Set Data

Now, let's create two sets of data 
- Large House: House which have more than 6 rooms.
- Small House: House which have less than 6 rooms.

Take sample of 100 house in each dataste.

```python
# ANSWER
N = 100
large_house = data[data['RM']>6]['MEDV'].sample(n=N, random_state=1)
small_house = data[data['RM']<=6]['MEDV'].sample(n=N, random_state=1)
```

##### 7. Calculate Variance

Calculate variance for both samples.

```python
a = large_house
b = small_house
```

```python
# ANSWER
large_house_var = large_house.var()
small_house_var = small_house.var()

print(f'Large House variance: {large_house_var}')
print(f'Small House variance: {small_house_var}')
```

##### 8. Calculate Standard Deviation


Because we are dealing with two samples we need to compute the joint standard deviation. 

Here sp is the pooled standard deviation for n = n1 = n2 and s^2
X1 and s^2 X2 are the unbiased estimators of the variances of the two samples.

![](https://wikimedia.org/api/rest_v1/media/math/render/svg/f63e99c30c01445668ea2cce9832da0f6810cb4b)

```python
# ANSWER
# Caculate the Standard Deviation
std = np.sqrt((large_house_var + small_house_var)/2)
```

```python
print('std dev: ', std)
```

##### 9. Calculate test statistic

The *t*-statistic depends on the difference between the sample means and their joint standard deviation:

![](https://wikimedia.org/api/rest_v1/media/math/render/svg/070d365e1b347ea5f83f0147043868fa120b6646)


```python
# ANSWER
tstat = ((a.mean() - b.mean())/(std*np.sqrt(2/N)))
tstat
```

##### 10. Calculate Degree of Freedom

The number of degrees of freedom is the total number of samples (here, this is N from each set) minus the number of statistics in the computation (1 for each sample mean):

```python
# ANSWER
df = 2*N -2
df
```

##### 11. Find the p-value

Now we can compute the probability of the above *t*-statistic from the cumulative density function:

> stats.t.cdf

> p_value = 1 - cdf(t) 

```python
# ANSWER
p = 1 - stats.t.cdf(tstat,df=df)
print(p)
print(str(abs(p/2)))
```

#### Interpret results

##### 12. Write a Function to Test Significance of `p_value`

Write a funtion which will take p_value and alpha as input. If p_value < alpha, print reject null hypothesis message. Otherwise print a fail to reject null hypothesis message. 

```python
# ANSWER
null_hypothesis = {'H0': "House price does not increase for large house.",
                  'H1': "House price increases for large house."}
```

```python
# ANSWER
def print_sig(p_value, alpha):
    if p_value < alpha: 
        print("We reject our null hypothesis.")
        print(null_hypothesis['H1'])
    elif p_value > alpha: 
        print("We fail to reject our null hypothesis.")
        print(null_hypothesis['H0'])
    else: 
        print("Our test is inconclusive")
```

```python
# ANSWER
print_sig(p, alpha)
```

##### 13. [Bonus] Write a function to calculate t, p_value from two samples.

Calculate the T-test for the means of *two independent* samples of scores.

This is a two-sided test for the null hypothesis that 2 independent samples
have identical average (expected) values.

```python
# ANSWER
def ttest_func(a, b, alpha=0.05):
    N = len(a)
    
    # variance
    a_var = a.var()
    b_var = b.var()
    print(f'{a} variance: {a_var}')
    print(f'{b} variance: {b_var}')
    
    # Caculate the Standard Deviation
    std = np.sqrt((a_var + b_var)/2)
    print('std dev: ', std)
    
    tstat = ((a.mean() - b.mean())/(std*np.sqrt(2/N)))
    print(tstat)
    
    df = 2*N -2
    print(df)

    p = 1 - stats.t.cdf(tstat,df=df)
    print(p)
    print(str(abs(p/2)))

```
