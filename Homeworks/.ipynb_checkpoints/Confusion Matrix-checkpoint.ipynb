```python
TP = 560
TN = 330
FP = 60
FN = 50
```

```python
# Accuracy
```

```python
accuracy = (TP + TN)/ (TP + TN + FP + FN)
print('Accuracy', accuracy)
```

```python
# Precision
```

```python
precision = TP / (TP + FP)
print('Precision (True positive rate)', precision)
```

```python
# Recall
```

```python
recall = TP / (TP + FN)
print('Recall', recall)
```
