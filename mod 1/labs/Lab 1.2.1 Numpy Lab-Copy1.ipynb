# Numpy Exploration

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# get matplot to show charts inline
%matplotlib inline
````

````python
height= np.round(np.random.normal(1.75, 0.20, 7000), 2)
weight= np.round(np.random.normal(60.32, 15, 7000), 2)
population=np.column_stack((height, weight))
````

````python
print('height:\n',height)
print('weight:\n',weight)
print('population:\n',population)
````

**1. Find Dimension of Array**

Find dimention of each array: height, weight, population
```
# Expected Ouput
np.ndim(height): 1
np.ndim(weight): 1
np.ndim(population): 2
```

````python
print("np.ndim(height):",np.ndim(height))
print("np.ndim(weight):",np.ndim(weight))
print("np.ndim(population):",np.ndim(population))
````

**8. Create a histogram of height and weight**

````python
help(plt.hist)
````

````python
num_bins = 50
n, bins, patches = plt.hist(height, bins=num_bins, facecolor='blue', alpha=.5)
plt.title("height")
plt.show()

plt.hist(weight, num_bins, facecolor='blue', alpha=0.5)
plt.title("weight")
plt.show()
````

**9. Create a boxplot of height and weight**

````python
help(plt.boxplot)
````

````python
plt.boxplot(height)
plt.yscale('linear')
# plt.grid(b=True, which='major', color='b', linestyle='.')
plt.title("height")
plt.show()

plt.boxplot(weight)
plt.title("weight")
plt.show()

plt.boxplot([height, weight], positions=[1,2])
plt.xticks([1,2], ["height","weight"])
plt.show()
````

**10. Create scatterplot of height and weight**

````python
help(plt.scatter)
````

````python
matplotlib.style.use('ggplot')
plt.scatter(weight, height)
plt.xlabel("weight")
plt.ylabel("height")
plt.show()
````
