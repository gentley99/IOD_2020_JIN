```python
import pandas as pd
from pytrends.request import TrendReq
```

```python
pytrend = TrendReq()
```

```python
pytrend.build_payload(kw_list=['bitcoin'])
```

```python
df = pytrend.interest_by_region()
df
```

```python
df.reset_index().plot(x='geoName', y='bitcoin', figsize=(30,10), kind='bar')
```

```python
#Get Google Hot Trends data
df = pytrend.trending_searches(pn='australia')
df.head()
```

```python
df
```

```python
df = pytrend.today_searches(pn='US')
```

```python
df
```

```python
# Get Google Keyword Suggestions
keywords = pytrend.suggestions(keyword='bitcoin')
df = pd.DataFrame(keywords)
df.drop(columns= 'mid')   # This column makes no sense
```
