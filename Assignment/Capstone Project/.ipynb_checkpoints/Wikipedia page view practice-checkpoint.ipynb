```python
import pageviewapi
import pandas as pd
import json
from attrdict import AttrDict
```

```python
all_views = pageviewapi.per_article('en.wikipedia', 'Bitcoin', '20210101', '20210525',
                        access='all-access', agent='all-agents', 
                        granularity='daily')
all_views
```

```python
for sub_attr in all_views: 
    print(all_views[sub_attr])
    all_views_1 = all_views[sub_attr]
```

```python
df = pd.DataFrame.from_dict(all_views_1, orient='columns')
df
```
