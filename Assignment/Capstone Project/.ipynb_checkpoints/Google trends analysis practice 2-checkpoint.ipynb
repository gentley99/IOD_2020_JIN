```python
import pandas as pd
from pytrends.request import TrendReq
```

```python
pytrends = TrendReq(hl='en-US')
```

```python
kw_list = ['bitcoin']
geo='US'
timeframes = ['today 5-y', 'today 12-m', 'today 3-m', 'today 1-m']
specific_tf = ['2021-01-01 2021-05-25']
gprop = ''
```

```python
pytrends.build_payload(kw_list, cat=0, timeframe='today 3-m', 
                       geo=geo, gprop=gprop)
```

# make sure to select the time carefully as the interest is relative to the highest point with in the data timeframe

```python
df = pytrends.interest_over_time()
df.head()
```

```python
df.bitcoin.mean()
```
