```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt

# Cause plots to be displayed in the notebook:
%pylab inline
%matplotlib inline
```

```python
df = pd.read_csv("Covid_GlobalEconomy.csv")
```

```python
# keeping columns with names that I can understand.
df1 = df[['iso_code', 'location', 'date', 'total_cases', 'total_deaths',
       'stringency_index', 'population', 'gdp_per_capita',
       'human_development_index']]
df1.head()
```

```python
df1.dtypes
```

```python
df1.info()
```

```python
## converting the date to date
df1['date'] = pd.to_datetime(df1['date'])
## set date as index
df1 = df1.set_index('date')
```

```python
df1.head()
```

```python
# exracting year
df1['year'] = pd.DatetimeIndex(df1.index).year
df1['year'].unique()
```

```python
# exracting month
df1['month'] = pd.DatetimeIndex(df1.index).month
df1['month'].sort_values().unique()
```

```python
# changing all values tofloat
df1['population'] = df1['population'].astype('float64')
df1.info()
```

```python
# checking missing values
df1.isnull().sum()
```

```python
# work with each one of the variables to deal with missing value
```

```python
# function to replace missing values as NaN
def missing_to_nan(data, column):
    missing = data[column].isna()
    data.loc[missing, column] = np.NaN

# function to replace missing values as zero
def missing_to_zero(data, column):
    missing = data[column].isna()
    data.loc[missing, column] = 0

# function to replace missing values using mean
def missing_to_mean(data, column):
    missing = data[column].isna()
    data.loc[missing, column] = data[column].mean()
```

```python
df1.describe()
```

```python
# I'm assuming that when there is missing values for total_cases and total_deathds,
# I will replace them to zero
missing_to_zero(df1, 'total_cases')
missing_to_zero(df1, 'total_deaths')
```

```python
df1.isnull().sum()
```

```python
df1.describe().T
```

```python
# to scale total_cases and total_deaths by population
df1['total_cases_scaled'] = df1['total_cases']/df1['population']*100
df1['total_deaths_scaled'] = df1['total_deaths']/df1['population']*100
df1.describe().T
```

```python
### pearson correlations as they are all continuous variables
df1.corr()
```

```python
# heatmap
sns.heatmap(df1.corr())
plt.title("Pearson correlation", size=20)
```

# Focusing on Australia, UK, US

```python
df1['location'].sort_values().unique()
```

```python
## Aus, UK (GBR), USA
aus_uk_usa = df1[(df1['location']=='Australia') 
                 | (df1['location']=='United Kingdom')
                 | (df1['location']=='United States')]
aus_uk_usa.head()
```

```python
aus = aus_uk_usa['location']=='Australia'
uk = aus_uk_usa['location']=='United Kingdom'
usa = aus_uk_usa['location']=='United States'
```

```python
## to scale so that we can compare with other countries
# scaled by population
```

```python
aus_uk_usa.info()
```

```python
aus_uk_usa.groupby(by=aus_uk_usa['iso_code']).describe().T
```

```python
aus_uk_usa.loc[aus].describe().T
```

```python
print(aus_uk_usa.loc[aus].index.duplicated().sum())
print(aus_uk_usa.loc[uk].index.duplicated().sum())
print(aus_uk_usa.loc[usa].index.duplicated().sum())
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, figsize=(10,10), sharex=True)

fig.suptitle("Total cases scaled by population", size=20)
ax1.plot(aus_uk_usa.loc[aus, 'total_cases_scaled'], label='Total cases')
ax1.plot(aus_uk_usa.loc[aus, 'total_deaths_scaled'], label='Total deaths')
ax2.plot(aus_uk_usa.loc[uk, 'total_cases_scaled'], label='Total cases')
ax2.plot(aus_uk_usa.loc[uk, 'total_deaths_scaled'], label='Total deaths')
ax3.plot(aus_uk_usa.loc[usa, 'total_cases_scaled'], label='Total cases')
ax3.plot(aus_uk_usa.loc[usa, 'total_deaths_scaled'], label='Total deaths')

ax1.set_xlabel('AUS')
ax2.set_xlabel('UK')
ax3.set_xlabel('USA')

plt.xticks(rotation=45)

plt.legend(loc="upper left")
```

```python
plt.plot(aus_uk_usa.loc[aus, 'total_cases_scaled'], label='AUS')
plt.plot(aus_uk_usa.loc[uk, 'total_cases_scaled'], label='UK')
plt.plot(aus_uk_usa.loc[usa, 'total_cases_scaled'], label='USA')

plt.title("Total cases")
plt.ylabel("Total cases scaled by population")
plt.xticks(rotation=45)

plt.legend(loc="upper left")
```
